```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
```

```python
x1 = np.linspace(-10, 10, 50)
y1 = np.sin(x1)

x2 = np.linspace(-10, 10, 1000)
y2 = np.cos(x2)

plt.title('Task # 4 Graph')
plt.xlabel("x axis")
plt.ylabel("y axis")

plt.plot(x1, y1,'o--', color="blue",  label="sin")
plt.plot(x2, y2,'--', color="red",  label="cos")
plt.legend(loc='upper right')


plt.xlim(-1, 5)
plt.ylim(-1, 1)

plt.savefig('data.png') 
plt.show()
```

# 4.4

```python
from math import log10
from numpy import exp
a=[]
b=[]
c=[]
d=[]
e1=[]
f1=[]
g=[]
h=[]


for x1 in range(-10,10,1):
    y1=x1**2+2*x1+2
    a.append(x1)
    b.append(y1)
             
for x2 in range(-10,10,1):
    y2=x2**3-x2**2+3*x2+2
    c.append(x2)
    d.append(y2)
    
for x3 in range(-5,20,1):
    y3=exp(x3)
    e1.append(x3)
    f1.append(y3)
    
for x4 in range(1,20,1):
    y4=log10(x4)
    g.append(x4)
    h.append(y4)

    
f, axarr = plt.subplots(2, 2)

axarr[0, 0].plot(a, b)
axarr[0, 0].set_title('$y=x^2+2x+2$')

axarr[0, 1].plot(c, d)
axarr[0, 1].set_title('$y=x^3-x^2+3x+2$')

axarr[1, 0].plot(e1,f1)
axarr[1, 0].set_title('$y=exp(x)$')

axarr[1, 1].plot(g, h)
axarr[1, 1].set_title('$y=log(x)$')
```

# 4.5

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd

X = np.linspace(0,5,51)
Y = np.linspace(0,5,51)

data = np.zeros((51, 51))

for i, x in enumerate(X):
    for j, y in enumerate(Y):
        data[i, j] = 3*x*y+x-2*y

df = pd.DataFrame(data, index=Y, columns=X)

sns.heatmap(df)
```
